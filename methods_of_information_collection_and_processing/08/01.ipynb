```python
import pandas as pd
import os
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('opendata.csv', encoding='cp1251')
```

```python
data_name = data['name'].unique()
for i in range(len(data_name)):
    print(f'{i}: {data_name[i]}')
action = int(input('Выберите номер действия: '))
```

```python
data_region = data['region'].unique()
for i in range(len(data_region)):
    print(f'{i}: {data_region[i]}')
region = int(input('Выберите регион: '))
```

```python
year = input('Введите год: ')
```

```python
pretty_year = f'за {year} год' if year else 'за всё время'
```

```python
user_data = \
    data[(data['name'] == data_name[action]) & (data['region'] == data_region[region]) & (data['date'].str.contains(year))]
```

```python
user_data.plot(kind='line',x='date',y='value',rot=45)
plt.xlabel("Дата")
plt.ylabel("Значение")
plt.title(f'{data_name[action]} по региону {data_region[region]} {pretty_year}')
plt.show()
```
