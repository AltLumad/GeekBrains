__2.__ Дан оператор поворота на 180 градусов, задаваемый матрицей 

$$A=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$$

Показать, что __любой__ вектор является для него собственным.   

##### Решение    
Пусть вектор $x=(k_1,k_2)$ , где $k_1,k_2)$ любые действительные числа    

$\begin{pmatrix}
-1 & 0\\ 
0 & -1\end{pmatrix}
\begin{pmatrix} k_1\\ k_2\end{pmatrix}=
\lambda\begin{pmatrix}k_1\\ k_2\end{pmatrix}$   

$\begin{cases}
-k_1=k_1\lambda \\ 
-k_2 = k_2\lambda
\end{cases}
\Rightarrow\lambda=-1$



__3.__ Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.



__Решение__   


$\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}
\begin{pmatrix}
1\\ 
1
\end{pmatrix}=
\lambda
\begin{pmatrix}
1\\ 
1
\end{pmatrix}.
$


$\begin{cases}
1+1 = \lambda \\ 
-1+3 = \lambda
\end{cases}
\Rightarrow
\lambda= 2$


$x=(1,1)$ является собственным вектором линейного оператора, заданного матрицей $A$.
