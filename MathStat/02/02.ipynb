##### 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.   


$P_n^k = C_n^kp^kq^{n-k}$

```python
from math import factorial as FA
C = 100*99*98*97*96*95*94*93*92*91*90*89*88*87*86/(FA(15)) #Есть опасения переполнениия на чисде 100! поэтому уже всё сокращено
P = C * 0.8**85 * 0.2**15
print(P)
```

##### 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?   


$p(k) = \frac{\lambda^k}{k!}e^{-\lambda}$

```python
import numpy as np
y=5000*0.0004
print('Вероятность, что ни одна лампочка не перегорит в первый день: ', 1-(y)*np.exp(-y))
print('Вероятность, что перегорят ровно две лампочки: ', (y**2)/2*np.exp(-y))
```

##### 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?   


```python
import numpy as np
C = np.math.factorial(144)/(np.math.factorial(70)*np.math.factorial(144-70))
P = C * 0.5**70 * 0.5**(144-70)
print(P)
```

##### 4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча.    
1. Какова вероятность того, что все мячи белые?    
2. Какова вероятность того, что ровно два мяча белые?    
3. Какова вероятность того, что хотя бы один мяч белый?   

```python
P1 = 7/10*6/9*9/11*8/10
print('1\t', P1)

P2  = 7/10*6/9*2/11*1/10 #два белых мяча из первого ящика
P2 += 7/10*3/9*9/11*2/10 + 3/10*7/9*2/11*9/10 #один белый из первого ящика и один белый из второго
P2 += 3/10*2/9*9/11*8/10 #два белых мяча из второго ящика
print('2\t', P2)

P3 = P1+P2
P3 += 7/10*6/9*9/11*2/10 + 7/10*6/9*2/11*9/10 #2 из первого, 1 из второго
P3 += 7/10*3/9*9/11*8/10 + 3/10*7/9*9/11*8/10 #1 из первого, 2 из второго
P3 += 7/10*3/9*2/11*1/10 + 3/10*6/9*2/11*1/10 + 3/10*2/9*9/11*2/10 + 3/10*2/9*2/11*9/10 #один из любого
print('3\t', P3)
```
