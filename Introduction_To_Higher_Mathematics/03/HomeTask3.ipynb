#### 1. Задание
###### Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10) .  Найдите их сумму. (на листочке)   


X = (10, 10, 10) + (0,0,-10) = (10, 10, 0)

###### Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
import math
def GetVectorLength(vec):
    s = 0
    for v in vec:
        s += v**2
    return math.sqrt(s)
```

```python
GetVectorLength([10, 10, 0])
```

#### 3. Задание (в программе)
###### Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.


```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
def DrawCircle(x0,y0, r):
    x = np.linspace(-r-1, r+1, 200)
    y = np.linspace(-r-1, r+1, 200)
    x, y = np.meshgrid(x, y)
    plt.figure(figsize=(6, 6))   
    plt.contour(x, y,((x-x0)**2/r**2 + (y-y0)**2/r**2), [1], colors='b')
    plt.show()
    
```

```python
DrawCircle(0.,0.,2.)
```

```python
def DrawElipse(a,b):
    x = np.linspace(-5, 5, 200)
    y = np.linspace(-5, 5, 200)
    x, y = np.meshgrid(x, y)
    plt.figure(figsize=(6, 6))   
    plt.contour(x, y,(x**2/a**2 + y**2/b**2), [1], colors='b')
    plt.show()
```

```python
DrawElipse(1,3)
```

```python
def DrawHyp(a,b):
    x = np.linspace(-5, 5, 200)
    y = np.linspace(-5, 5, 200)
    x, y = np.meshgrid(x, y)
    plt.figure(figsize=(6, 6))   
    plt.contour(x, y,(x**2/a**2 - y**2/b**2), [1], colors='b')
    plt.show()
    
```

```python
DrawHyp(1.,2.)
```

```python
def Polar2Decart(R,a):
    x = R*np.cos(a)
    y = R*np.sin(a)
    return (x,y)
```

```python
def DrawPolarCircle(R, a):
    x,y = Polar2Decart(R,a)
    DrawCircle(x,y, R)
    
```

```python
DrawPolarCircle(1,1)
```

```python
def DrawPolarLine(R1, a1, R2, a2):
    x1,y1 = Polar2Decart(R1,a1)
    x2,y2 = Polar2Decart(R2,a2)
    x = np.linspace(min(x1,x2),max(x1,x2), 200)
    y = np.linspace(min(y1,y2),max(y1,y2), 200)
    plt.plot(x, y)
    
```

```python
DrawPolarLine(1,1, 2, 4)
```

4. Задание (в программе)   
 - Решите систему уравнений:   
y = $x^2 – 1$   
exp(x) + x∙(1 – y) = 1   



```python
x = np.linspace(-5, 5, 300)
plt.plot(x, (x**2 - 1))
plt.plot(x, (np.exp(x) + x  - 1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def f(p):
    x,y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)

x1, y1 = fsolve(f,(-2.,2.))
print(x1, y1)
```
